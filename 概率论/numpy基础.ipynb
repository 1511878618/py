```python
test1 = np.random.rand(3,4,5)
print('未调用随机数种子时候',test1)
np.random.seed(5)
test2 = np.random.rand(3,4,5)
print('调用随机数种子后',test2)
np.random.seed(5)
test3 = np.random.rand(3,4,5)
print('再一次调用随机数种子的结果',test3)
```

```python
np.random.seed()
a = np.arange(100).reshape(4,5,5)
print(a)
print('调用permutation后\n')
b = np.random.permutation(a)
print('b是随机打乱后的数组{} \n a是初始数组 {}'.format(b,a))
print('调用shuffle后\n')
np.random.shuffle(a)
print('a是初始数组 {}'.format(a))
```
