```python
import numpy as np 
import math 
```

```python
def sin(jiaodu):
    return math.sin(np.pi*jiaodu / 180)

def cos(jiaodu):

    return math.cos(np.pi*jiaodu / 180)
```

```python
class puzzle(object):
    
    

    def __init__(self,L):
        self.L = L
        self.Fm = None
        self.a = None
        self.R =None
        
    def SolveFm(self):
        L = self.L 
        Na = (3*10/(30+L))
        Fm = (4.8*Na +0.05)/(7*sin(70)-1.9*cos(70))
        self.Fm = Fm 
        return Fm

    def Solvea(self):
        Fm = self.Fm 
        a = np.arctan((Fm*sin(70) - 1/6 +(30/(30+self.L)))/(Fm*cos(70)))*180/np.pi
        self.a = a 
        return a
    def SolveR(self):
        R = self.Fm*cos(70)/cos(self.a)
        self.R = R 
        return R 
    def __str__(self):
        self.SolveFm()
        self.Solvea()
        self.SolveR()
        return '在L为{}cm时：\nFm为{:.2f}N\n对应得a得角度为：\n{:2f}°\n以及R得大小为：\n{:.2f}N'.format(self.L,self.Fm,self.a,self.R)
        
```
