# 下载网页
爬取的第一步便是，通过requests下载网页。
其主要内容包括：
1. headers的定制（"User-Agent等"）

常用的`"User-Agent":"user-agent: Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/80.0.3987.163 Safari/537.36"`
2. 请求网页

`requests.get(url,headers=headers`

3. 返回`r.text`，即网页的html源码，供给网页解析库使用。如：Beautiful Soup

```python
def download_page(url):
    headers= {"User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/80.0.3987.163 Safari/537.36"}
    r = requests.get(url,headers=headers)
    print(r.encoding)
    return r.text 
    
```

# 获取总的页数
获取段子的页数，以保证爬取到了所有的段子


```python
def get_pages():
    url = 'https://www.qiushibaike.com/text/'
    html = download_page(url)
    Soup = BeautifulSoup(html,'html.parser')
    Pages = Soup.find_all('span',class_='page-numbers')[-1]
    sum_pages= int(Pages.text)
    return sum_pages
    
```

# 获取网页中所需的内容
通过对网页的HTML结构进行解析。

使用Beautiful Soup对网页进行解析

大体步骤：
1. `soup = BeautifulSoup(html,'html.parser')`

2. ` xx = soup.find()`对所需要的html内容通过标签进行检索。


```python
def get_content(html,page):
    #定义输出的格式
    output = """第{}页 作者：{} 性别：{} 年龄：{} 点赞：{} 评论：{}\n {}\n ------\n"""
    Soup = BeautifulSoup(html,'html.parser')
    # 爬取所需的内容
    Cont_list = Soup.find_all('div',class_=re.compile('article block untagged mb15'))
    for cont in Cont_list:
        author = cont.find('h2').text
        sex = cont.find('div',class_ = re.compile('articleGender'))['class'][1]
        if sex == 'manIcon':
            sex = 'man'
        if sex == 'womenIcon':
            sex = 'women'
        else:
            sex ='unknown'
            
        age = cont.find('div',class_ = re.compile('articleGender')).text
        dianzan = cont.find('span',class_ ='stats-vote').find('i').text
        comments = cont.find('span',class_='stats-comments').find('i',class_='number').text
        contents = cont.find('div',class_='content').text
        save_text(output.format(page,author,sex,age,dianzan,comments,contents))
        
```

# 保存爬取的内容
创建一个可以重复写入的文档

注意使用`'a'`


```python
def save_text(text):
    for i in text:
        with open('qiubai.txt','a',encoding='utf-8')as f:
            f.write(i)
```

#  写主程序


```python
def main():
    for i in range(1,get_pages()+1):
        url = 'https://qiushibaike.com/text/page/{}'.format(i)
        html = download_page(url)
        get_content(html,i)
        
if __name__=='__main__':
    main()
```
