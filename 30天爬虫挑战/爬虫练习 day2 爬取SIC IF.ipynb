```python
def write_text(text):
    with open('qikan.txt','a') as f:
        f.write(text)
```

```python
def get_content(response):
    response = response['data']
    if response:
        for i in response: 
            html = i['content']
            Soup = BeautifulSoup(html,'html.parser')
            Article = Soup.find('a',title='去查看期刊的详情页').text
            Content = Soup.find('span',class_='layui-breadcrumb').text.split('\n')
            for i in Content:
                if re.match(re.compile('研究方向'),i):
                    Research = i 
                if re.match(re.compile('投稿周期'),i):
                    Time = i 
                else:
                    pass
            output = '期刊名称：{}\n研究方向：{}\n投稿周期：{}\n ——————\n'.format(Article,Research,Time)
            write_text(output)
        

        
    else:
        print('没有数据啦')
        return '没有数据啦'
        
```
