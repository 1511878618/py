```python
import requests
from bs4 import BeautifulSoup
import re 
import time 
```

```python
def download_url(url):
    headers = {
  'Referer': 'https://www.biqukan.com/',
  'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/81.0.4044.113 Safari/537.36'
}
    r = requests.get(url,headers=headers)
    r = r.content.decode(encoding='GBK',errors="ignore")
    return r
```

```python
def get_bookContent(book_html):
    time.sleep(0.1)
    Soup = BeautifulSoup(book_html,'html.parser')
    bookName = re.search(r'《.*》',Soup.dt.text ).group(0)
    charpters = Soup.find_all('dd')
    TotalCharpterNum = len(charpters)
    print(TotalCharpterNum)
    index = 1 
    for dd in charpters:
        refer = 'https://www.biqukan.com'
        href = refer + dd.a['href']
        charpter = dd.text
        html = download_url(href)
        soup = BeautifulSoup(html,'html.parser')
        content = soup.find('div',class_='content').text 
        with open(bookName+'.txt','a',encoding='utf-8') as f:
            f.write(content+'\n')
            print('已经下载{:.2f}%\n'.format(index/TotalCharpterNum))
            index +=1 
```

```python
def main():
    
    book_url =input('请输入小果果你想要下载的小说的url：')
    book_html = download_url(book_url)
    get_bookContent(book_html)
    print('下载完成哒哒哒')
```

```python
main()
```
